```python
df["Total Cost"] = df['quantity'] * df['cost_price_per_unit']
```

```python
df["Total Profit"] = df["Total Sales"] - df["Total Cost"] 
```

#QUESTION NO 13

```python
category_sales = df.groupby('category')['Total Sales'].sum()
category_profit = df.groupby('category')['Total Profit'].sum()

profit_margin = (category_profit / category_sales) * 100

sorted_profit_margin = profit_margin.sort_values(ascending=False)

sorted_profit_margin.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Profit Margins by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin (%)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
