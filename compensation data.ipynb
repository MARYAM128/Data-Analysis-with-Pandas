```python
for i in survey:
    years = i[0]
    if years < 5:
        category = "< 5 years"
    elif years < 10:
        category = " 5 - 10 years"
    elif years < 15:
        category = " 10 - 15 years"
    elif years < 20:
        category = " 15 - 20 years"
    elif years < 25:
        category = "< 20 - 25 years"
    else:
        category =  "25+ years" 

    i.append(category)
    
```

```python
lessthan5 = []
fiveto10 = []
tento15 =[]
fifteento20 =[]
twentyto25 =[]
morethan25 =[]

for i in survey:
    if i[-1] == "< 5 years":
        lessthan5.append(i[-2])
    elif i[-1] == " 5 - 10 years":
        fiveto10.append(i[-2])
    elif i[-1] == " 10 - 15 years":
        tento15.append(i[-2])
    elif i[-1] == " 15 - 20 years":
        fifteento20.append(i[-2])
    elif i[-1] == "< 20 - 25 years":
        twentyto25.append(i[-2])
    else :
        morethan25.append(i[-2])
        
```

```python
cat= [lessthan5 ,fiveto10, tento15, fifteento20, twentyto25,morethan25]
for c in cat:
    print(sum(c)/len(c))
```
