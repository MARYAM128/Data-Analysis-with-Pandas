```python
import pandas  as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("cars.csv")
df.head()
```

```python
df
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df["Vehicle Size"].unique()
```

```python
df = df.drop(["Vehicle Size"], axis = 1)
df.head()
```

```python
df = df.rename(columns={"Engine HP" : "HP", "Engine Cylinders" : "Cylinders", "Transmission Type" : "Transmission" , "Driven_Wheels" : "Drive Mode", "highway MPG" : "MPG", "city mpg" : "MPG-C", "MSRP" : "Price"}) 
df.head()
```

```python
df.shape
```

```python
df.duplicated()
```

```python
df[df.duplicated()]
```

```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ",duplicate_rows_df.shape) 
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```
